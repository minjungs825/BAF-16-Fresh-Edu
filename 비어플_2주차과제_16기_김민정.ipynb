Pregnancies : 임신 횟수

## 로지스틱 회귀모델

```python
data.head(3)
```

```python
#outcome 값 확인
data['Outcome'].value_counts()
```

```python
#학습, 예측 데이터셋 나누기- 80%학습 데이터
split_count =int(data.shape[0]*0.8)
split_count
```

```python
#힉습데이터
train=data[:split_count].copy()
train
```

```python
#훈련 데이터
test=data[split_count:].copy()
test
```

```python
#outcome만 제거하고 나열
feature_names=train.columns[:-1].tolist() 
feature_names
```

```python
#예측해야될 값-outcome
label_name=train.columns[-1]
label_name
```

```python
#학습, 예측 데이터셋 만들기
#train 데이터셋 (8개 칼럼)
X_train=train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
#train 데이터셋 (1개 칼럼)
y_train=train[label_name]
y_train.head()
```

```python
#test 데이터셋 (8개 칼럼)
X_test=test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
#test 데이터셋 (1개 칼럼)
y_test=test[label_name]
y_test

#X_train, X_test, y_train, y_test = train_test_split()을 통해서도 분할 가능
```

```python
#로지스틱 회귀 모델 학습
clf = LogisticRegression(max_iter=1000) #반복 횟수
clf.fit(X_train, y_train)
```

```python
#예측
y_pred = clf.predict(X_test)
y_pred
```

```python
#혼동 행렬: 예측 결과를 실제값과 비교
print(confusion_matrix(y_test, y_pred))
#TN=70, FP=9, FN=23, TP=17
```

```python
#정확도/ 재현율/ f1스코어
print(classification_report(y_test, y_pred))
```

```python
#피처 중요도 계산
feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': clf.coef_[0]
})
feature_importance['Importance'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
print("\nFeature Importance:")
print(feature_importance)
#DiabetesPedigreeFunction가 높은 중요도를 가짐
```

```python
#피처 중요도 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title('Feature Importance (Logistic Regression)')
plt.show()
```
