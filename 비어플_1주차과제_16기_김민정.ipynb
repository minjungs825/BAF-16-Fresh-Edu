#### 변수별로 특징 확인하기

1. 등급별 생존여부 확인

```python
df2['Pclass'].value_counts()
```

```python
#등급별 생존의 평균 시각화
sns.barplot(data=df2,x='Pclass',y='Survived')
#1등급은 상대적으로 높은 생존을 하였지만 3등급은 낮은 생존율을 보인다.
```

```python
# 성별 추가
#등급대별 성별을 나눠 생존율을 분석/시각화
sns.barplot(data=df2, x='Pclass',y='Survived', hue='Sex')
#대체적으로 3등급 모두 다 female이 더 높은 생존율을 가진다.
```

```python
#탑승 항구에 따른 생존율 분석/시각화
sns.barplot(data=df2, x='Embarked',y='Survived')
#C 탑승구에서 승차한 사람들의 생존율이 더 높음
```

```python
#나이 결측값 전처리
df2['Age']=df2['Age'].dropna()
```

```python
#Cabin도 많은 결측치가 존재하므로 결측치를 제거한다.
df2['Cabin']=df2['Cabin'].dropna()
df2['Cabin'].count()
```

```python
#나이 분포 확인sns.histplot(df2['Age'],kde=True,bins=30)
#20대~30대가 많이 승차했음을 알 수 있다. 
```

```python
#생존 여부에 따른 형제자매/배우자 수
sns.countplot(data=df2, x='SibSp', hue='Survived')
#대부분 혼자가 많이 승차했지만, 약 400명이 사망했다는걸 알 수 있다.
```

```python
# 각 변수에 대한 상관관계를 히트맵으로 시각화
sns.heatmap(df2.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("상관관계 히트맵")
plt.show()
```
